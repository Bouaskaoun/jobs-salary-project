```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/MedOvich/IdeaProjects/jobs-salary-project/data/glassdoor_jobs_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[~df['Salary Estimate'].isna()]
len(df)
```

```python
salary_count = df.pivot_table(columns=['Salary Estimate'], aggfunc='size')
salary_count
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer est.' in x.lower() else 0)
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.lower().split('(')[0].replace('k','').replace('$','').replace('per hour',''))
salary
```

```python
df[["min_salary", "max_salary"]] = salary.str.split("-").apply(pd.Series).ffill(axis=1)
df['min_salary'] = df['min_salary'].apply(lambda x: int(x))
df['max_salary'] = df['max_salary'].apply(lambda x: int(x))
df['avg_salary'] = (df['min_salary'] + df['max_salary'])/2
```

```python
df.head()
```

```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split()[0])
```

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')).apply(lambda x: x[1] if len(x)>1 else None)
```

```python
df
```

```python
df.info()
```

```python
df['age'] = df['Founded'].apply(lambda x: x if np.isnan(x) else 2022 - x)
```

```python
df['Python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['SQL'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df['PowerBI'] = df['Job Description'].apply(lambda x: 1 if 'powerbi' in x.lower() else 0)
df['Tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['AWS'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
```

```python
df.head()
```

```python
df = df.iloc[: , 1:]
```
